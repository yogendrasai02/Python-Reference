# Variables

- A variable is a named memory location. 

- A variable can hold data of various types, referred to as 'Data Types'.

- In Python, you can create & assign a variable, some value using the following syntax:
`variable_name = value_of_variable`

- Here, `=` is called the **Assignment Operator**

- We must follow some rules while naming variables.

## Variable (or Identifier) Naming Rules

- Identifier Names can contain only alphabets, digits, or underscore (`_`).

- Names must begin with either an alphabet or an underscore.

- Names cannot contain any spaces, or any other special characters other than underscore.

- Names must not use a Python Keyword, such as `list, set, sum, max, min` etc.

- According to [PEP8 convention](https://peps.python.org/pep-0008/#function-and-variable-names), it is recommended to keep the names in lowercase, with words separated by `_` if needed.

- Examples of valid Variable|Identifier names: `my_tax, salaries, pricesData, myNum1, x1, y_1, message_text_body`

- Examples of invalid Variable names: `0001, 007, 0myDog, myDog is dope`

- Python uses **Dynamic Typing**. It means that a variable can be reassigned with a value of different datatype. See the below code block for an example.

```python
# Dynamic Typing

# String
names = "Tucker, Todd"
print(names, type(names))

# List
names = ["Tucker", "Todd"]
print(names, type(names))
```

> `print()` is an inbuilt python function to write some data to the standard output. It accepts String data but typecasts other types to string while printing.

> `type()` is an inbuilt python function which returns the **Data type** of the variable which was passed as an argument.

>  Dynamic Typing feature in Python is something which we need to be careful and aware about. It might lead to unexpected errors or bugs.
