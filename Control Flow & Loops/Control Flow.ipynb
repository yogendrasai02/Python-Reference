# Control Flow in Python

- Control Flow statements allow us to conditionally execute code, based on a logical condition.

- These let you tell the interpretor statements like: *If tank is empty, then fill the tank!*.

- Three keywords are used:
    - `if` keyword
    - `elif` keyword
    - `else` keyword
    
- For the control flow statements, not the `1 TAB Space` **indentation**.

## `if` keyword

- Syntax for `if` statement:

```python
if some_condition:
    ...code to execute...
```

## `else` keyword

- If none of the conditions in `if` or `elif` blocks are `True`, then the statements under `else` block are executed.

- The use of `else` block is optional.

- Syntax:

```python
if condition_1:
    ...code...
else:
    ...code...
```

- Here a flow chart:


## `elif` keyword

- `elif` is basically like `else if` statements.

- Syntax is:

```python
if condition_1:
    ...code...
elif condition_2:
    ...code...
elif condition_3:
    ...code...
..
..
..
else: # optional
    ...code...
```

- Heres a flow chart:


````python
# if keyword
isTired = True

if isTired:
    print('Take Rest')
print('Done')

# Below code is analogous to above statement
if isTired == True:
    print('Take Rest')
print('Done')
    
if not isTired:
    print('Work!!')
print('Done')    
````

````python
# if-else

isFull = False

if isFull:
    print('Tank is full!')
else:
    print('Tank is not full!')
````

````python
# elif ladder

income = 1000
tax = None

if income < 200:
    tax = 0
elif income >= 200 and income < 500:
    tax = 0.2 * income
elif income >= 500 and income < 800:
    tax = 0.25 * income
else:
    tax = 0.3 * income
print(f'Your tax is {tax} for income {income}')
````
