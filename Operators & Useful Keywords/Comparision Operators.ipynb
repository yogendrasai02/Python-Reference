# Comparision Operators

- These operators take two operands and compare their value, and return `True` if the comparision holds, else return `False`.

The following are the comparision operators:



(Reference: [GFG](https://www.geeksforgeeks.org/python-operators/))

### `==` operator

```python
# == operator check for equality
10 == 10
```

```python
# With floating point, it returns true if the values are same
10 == 10.0
```

```python
10 == 5
```

```python
'hello' == 'hello'
```

```python
'hello' == "hello"
```

```python
# note that with strings, it is CASE SENSITIVE
'hello' == 'Hello'
```

### `>` operator and `>=` operator

```python
10 > 5
```

```python
5 > 10
```

```python
10 >= 5
```

```python
5 >= 10
```

```python
# Again, value is checked
10 >= 10.0
```

```python
# With strings, LEXICOGRAPHICAL COMPARISION is done
'a' > 'b'
```

```python
'sheldon' > 'leonard'
```

```python
'pick' > 'pack'
```

```python
'light' >= 'lite'
```

```python
'lite' >= 'light'
```

### `<` operator and `<=` operator

```python
10 < 5
```

```python
5 < 10
```

```python
10 <= 5
```

```python
5 <= 10
```

```python
# Again, with strings, LEXICOGRAPHICAL COMPARISION IS DONE
'bag' < 'beg'
```

```python
'bag' < 'Bag'
```

```python
'Bag' < 'bag'
```

### `!=` operator

```python
10 != 5
```

```python
10 != 10.0
```

```python
'hello' != 'hello'
```

```python
'hello' != 'Hello'
```
