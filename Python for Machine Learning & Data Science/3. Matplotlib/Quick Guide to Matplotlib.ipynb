## `matplotlib.pyplot`

- `matplotlib.pyplot` is a collection of functions that make matplotlib work like MATLAB. 

- Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

- Using this module requires less code, but it is not as flexible as the core matplotlib is.

- Plotting functions expect `numpy.array` as input, or objects that can be passed to `numpy.asarray`. Common convention is to convert these to `numpy.array` objects prior to plotting.

- We will now see how to plot some useful visualisations.

````python
# importing required libraries
import matplotlib.pyplot as plt
import numpy as np
````

## `plt.plot()` - To plot 2D data as Line Chart / Using Markers

- The `plot()` function is used to plot X-Y 2D data, in the form of lines, or by using custom markers.

- Syntax: `matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)`

- `*args` argument:

    - `x` array which denotes X coordinates. (optional)
    
    - `y` array which denotes Y coordinates. (required)
    
    - `fmt` denoting a format string, to quickly setup the basic line properties. (optional)
    
        - Syntax for `fmt` is `[marker_short][line_short][color_short]`
        
        - Some examples: 
```python   
'b'    # blue markers with default shape
'or'   # red circles
'-g'   # green solid line
'--'   # dashed line with default color
'^k:'  # black triangle_up markers connected by a dotted line
```    

- `**kwargs` argument specify the properties of the line. They are optional. Some useful ones are listed below:

    - `color` or `c`: sets the line color.
        
        - `b` or `blue`
        
        - `green`, `red`, `cyan`, `magenta`, `yellow`, `black`, `white`
    
        - [Visit this for more information](https://matplotlib.org/stable/api/colors_api.html#module-matplotlib.colors)
    
    - `linestyle` or `ls`: sets the linestyle and can be on the following:
    
        - `-` or `solid`
        
        - `--` or `dashed`
        
        - `-.` or `dashdot`
        
        - `:` or `dotted`
        
    - `linewidth` or `lw`: pass in a `float`, which sets the line width in points.
    
    - `marker`: used to define the marker style. [List of marker styles](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers)

- In the below example, lets plot a simple line chart.

- `plt.title()` sets the title for the chart.

- `plt.xlabel()` sets the X-Axis label. Similarly, we have `plt.ylabel()`.

- `plt.show()` outputs the actual graph plotted.

- A common convention is to convert the inputs into `numpy.array`.

````python
# simple line-chart
size = np.array([10, 20, 30, 40, 50])
freq = np.array([5, 7, 6, 8, 4])

plt.plot(size, freq)

plt.title('Size vs Customers')
plt.xlabel('Size')
plt.ylabel('Number of Customers')

plt.show()
````

- We can use `plt.figure(figsize = (X_Inches, Y_Inches))` to plot a figure with custom dimensions.

````python
# custom figure size
# 10 inches X 5 inches
fig = plt.figure(figsize = (10, 5))

plt.plot(size, freq)
plt.show()
````

- We can plot multiple datasets in a single plot.

- We can also set a legend for the plot.

````python
# multiple lines in a single plot & legend
from random import randint

ratings_jan = [i for i in range(1, 11)]
customers_jan = [randint(1, 100) for i in range(1, 11)]

ratings_feb = [i for i in range(1, 11)]
customers_feb = [randint(1, 100) for i in range(1, 11)]

plt.plot(ratings_jan, customers_jan, color='green')
plt.plot(ratings_feb, customers_feb, color='red')

plt.xlabel('Rating')
plt.ylabel('Number of Customers')
plt.title('Ratings v/s Customers')

plt.legend(['Jan', 'Feb'])

plt.show()
````

- lets customize our default plot.

````python
# color, linewidth, linestyle, marker
house_size = np.array([500, 900, 1000, 1400, 1450, 1500])
house_prices = np.array([randint(1000, 10000) for i in range(len(house_size))])

plt.plot(house_size, house_prices, 
         color='red', marker='o', markerfacecolor='orange', markersize=10, 
         linestyle='dashed', linewidth=3)

plt.xlabel('House Size')
plt.ylabel('House Price')

plt.show()
````

````python
# using a format string
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([3, 2, 1, 5, 6, 4])

plt.plot(x, y, 'X:g', markersize=10, linewidth=3, markeredgecolor='r', markerfacecolor='r')

plt.show()
````

````python
# plotting categorical variables

animals = np.array(['panda', 'dog', 'cheetah', 'koala'])
population = np.array([1000, 9000, 1500, 5000])

plt.plot(animals, population, marker='*', ls='', markersize=10)
plt.show()
````

````python
# using a grid
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([3, 2, 1, 5, 6, 4])

plt.plot(x, y, 'X-g', markersize=10, markerfacecolor='r', markeredgecolor='r')
plt.grid(True)

plt.show()
````

### Multiple Plots - Subplots

````python
# plotting sine, cosine, tangent, log functions

import math

# data generation
angles = np.arange(0, 2 * math.pi, 0.05)
sine_vals = np.sin(angles)
cosine_vals = np.cos(angles)
tangent_vals = np.tan(angles)
log_x = np.arange(0, 300)
log_vals = np.log(log_x)

# create a big fig
fig = plt.figure(figsize=(10, 10))

# Create a 2X2 matrix
# in that matrix, plot at location 1

plt.subplot(2, 2, 1)
plt.plot(angles, sine_vals)
plt.title('Sine')
plt.grid(True)

# in that matrix, plot at location 1
plt.subplot(2, 2, 2)
plt.plot(angles, cosine_vals)
plt.title('Cosine')
plt.grid(True)

# in that matrix, plot at location 1
plt.subplot(2, 2, 3)
plt.plot(angles, tangent_vals)
plt.title('Tangent')
plt.grid(True)

# in that matrix, plot at location 1
plt.subplot(2, 2, 4)
plt.plot(log_x, log_vals)
plt.title('log')
plt.grid(True)

plt.show()
````

## `plt.scatter()` - Creating Scatter Plots

- A scatter plot (also called a scatterplot, scatter graph, scatter chart, scattergram, or scatter diagram) is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data.

- Syntax: `matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)`

- `x` denotes the X coordinates array and `y` denotes the Y coordinates array.

````python
# Creating a simple scatter plot
from random import randint

size = np.array([randint(1, 100) for i in range(10)])
price = np.array([randint(1, 100) for i in range(10)])

plt.scatter(size, price, c='r', marker='+')

plt.title('House Size v/s House Price')
plt.xlabel('House Size in 1000 sq.ft')
plt.ylabel('House Price in $100k')

plt.show()
````

````python
# Plotting two datasets in a single scatter plot
size_delhi = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
price_delhi = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

size_hyd = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
price_hyd = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(size_delhi, price_delhi)
plt.scatter(size_hyd, price_hyd)

plt.show()
````

## `plt.bar()` - To plot Bar graphs

- Bar graphs are the pictorial representation of data (generally grouped), in the form of vertical or horizontal rectangular bars, where the length of bars are proportional to the measure of data. 

- They are also known as bar charts. Bar graphs are one of the means of data handling in statistics.

- Syntax: `matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)`

- `x` denotes the X coordinates array.

- `height` denotes the height of bars.

````python
# default bar graph
space_org = np.array(['ISRO', 'NASA', 'SpaceX', 'ESA', 'Blue Origin', 'Virgin Galactic'])
launches = np.array([10, 25, 29, 15, 4, 2])

plt.figure(figsize=(10,5))

plt.bar(space_org, launches)
plt.xlabel('Space Organization')
plt.ylabel('Number of Launches')

plt.show()
````

````python
# multiple bars
plt.figure(figsize=(10,5))

launches_2021 = np.array([12, 30, 42, 12, 6, 3])
launches_2022 = np.array([20, 35, 49, 10, 4, 2])

x_coord = np.arange(len(space_org))

plt.xticks(x_coord, space_org)
plt.bar(x_coord - 0.2, launches_2022, width = 0.4)
plt.bar(x_coord + 0.2, launches_2021, width = 0.4)

plt.legend(['Year 2022', 'Year 2021'])

plt.xlabel('Space Organization')
plt.ylabel('Number of Launches')

plt.show()
````

## `plt.hist()` - To plot Histograms

- A histogram is a graph showing *frequency* distributions.

- It is a graph showing the number of observations within each given interval.

- Syntax: `matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs`

- `x` denotes the input values.

- `bins` represent the interval size. (default 10)

    - If it is an integer, it represents equal width intervals.
    
    - If it is a sequence, it represents the edges of the intervals.
    
- The input values are divided as per the bins, and the frequency bars are plotted in the histogram.

````python
# Plotting a simple, default histogram
voters_age = np.array([randint(18, 100) for i in range(25)])

fig = plt.figure(figsize=(10, 5))

plt.hist(voters_age)

plt.title('Voters age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
````

## `plt.pie()` - Plotting a Pie chart

- A pie chart (or a circle chart) is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. 

- In a pie chart, the arc length of each slice (and consequently its central angle and area) is proportional to the quantity it represents. 

- Syntax: `matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, normalize=True, data=None)`

- Make a pie chart of array `x`. The fractional area of each wedge is given by `x/sum(x)`.\

- `explode`: If not `None`, is a `len(x)` array which specifies the fraction of the radius with which to offset each wedge.

- `labels`: a sequence of string to label each wedge. (specify as a keyword argument)

- `color`: sequence of colors for each wedge.

````python
# simple, default pie chart
expenses = [1000, 500, 720, 900, 50, 100, 1500]
labels = ['fuel', 'diary products', 'food', 'gym', 'newspaper', 'tips', 'college']

fig = plt.figure(figsize = (10, 10))

plt.pie(expenses, labels = labels)

plt.axis('equal')

plt.show()
````

````python
# show percentages as well
expenses = [1000, 500, 720, 900, 50, 100, 1500]
labels = ['fuel', 'diary products', 'food', 'gym', 'newspaper', 'tips', 'college']

fig = plt.figure(figsize = (10, 10))

plt.pie(expenses, labels = labels, autopct='%1.1f%%')

plt.axis('equal')

plt.show()
````
