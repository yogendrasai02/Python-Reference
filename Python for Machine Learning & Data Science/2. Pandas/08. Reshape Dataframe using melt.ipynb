# Reshaping dataframe using `df.melt()`

- `melt` is basically opposite of `pivot` &rarr; Unpivot a dataframe.



- `id_vars`: index

- Consider the following dataframe:

```python
import pandas as pd

df = pd.read_csv('data/weather_melt.csv')
df
```

```python
# melt
df.melt(id_vars=['day'])
```

```python
# melt
df.melt(id_vars=['day'], var_name='city', value_name='temperature')
```
